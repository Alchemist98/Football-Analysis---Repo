```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_excel("ligue1.xlsx")
data.head()
```

```python
attackers = data[data['Pos'] == 'FW']  
```

```python
attackers
```

```python
attackers[attackers['Age']<=26].head()
```

```python
attackers.columns
```

```python
attackers.reset_index(drop=True,inplace=True)
```

```python
attackers
```

```python
attacker = attackers[attackers['Gls'] >=10]
attacker.reset_index(drop=True,inplace=True)
```

```python
attacker
```

```python
fig = plt.figure()
fig.set_size_inches(10,7)
ax = fig.add_subplot(1,1,1)

plt.scatter(attacker['Gls.1'],attacker['Min'],color='green')
plt.xlabel("Goals per 90s",size=16)
plt.ylabel("Total minutes played",size=16)
plt.suptitle("Ligue Best Strikers",size=16,y=0.93)

for i,txt in enumerate(attacker['Player']):
    ax.annotate(txt,(attacker['Gls.1'][i],attacker['Min'][i]))
```
