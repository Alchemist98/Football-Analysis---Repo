```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_excel("barclays.xlsx")
data.head()
```

```python
data.columns
```

```python
defenders = data[data['Pos'] == 'DF']
defenders
```

```python
defe = defenders[defenders['90s'] >= 22]
defe
```

```python
len(defe)
```

```python
defe.reset_index(drop=True,inplace=True)
```

```python
defe.head()
```

```python
fig = plt.figure()
fig.set_size_inches(15,10)
ax = fig.add_subplot(1,1,1)
plt.style.use('grayscale')

plt.scatter(defe['Def 3rd'],defe['90s'],color='blue')
plt.xlabel("Defensive tackles in defensive 3rd",size=16)
plt.ylabel("Number of 90s played",size=16)

for i,txt in enumerate(defe['Player']):
    ax.annotate(txt,(defe['Def 3rd'][i],defe['90s'][i]))
```

```python
import seaborn as sns
```

```python
best_mid_3rd_def = defe.sort_values(by='Mid 3rd',ascending=False)
```

```python
best_mid_3rd_def.reset_index(drop=True,inplace=True)
```

```python
df = best_mid_3rd_def[:30]
```

```python
df
```

```python
fig = plt.figure()
fig.set_size_inches(15,10)
ax = fig.add_subplot(1,1,1)
plt.style.use('seaborn-whitegrid')

#sns.stripplot(df['Mid 3rd'],df['90s'],jitter=0.25, size=8, ax=ax, linewidth=.5)
plt.scatter(df['Mid 3rd'],df['90s'],color='blue')
plt.xlabel("Defensive tackles in middle 3rd",size=16)
plt.ylabel("Number of 90s played",size=16)
plt.suptitle("Best tacklers in middle 3rd",size=20,y=0.92)

for i,txt in enumerate(df['Player']):
    ax.annotate(txt,(df['Mid 3rd'][i],df['90s'][i]))
```

```python
fig = plt.figure()
fig.set_size_inches(15,10)
ax = fig.add_subplot(1,1,1)
plt.style.use('dark_background')

#sns.stripplot(df['Mid 3rd'],df['90s'],jitter=0.25, size=8, ax=ax, linewidth=.5)
plt.scatter(df['Att 3rd'],df['90s'],color='blue')
plt.xlabel("Defensive tackles in attacking 3rd",size=16)
plt.ylabel("Number of 90s played",size=16)
plt.suptitle("Best tacklers in attacking 3rd",size=20,y=0.92)

for i,txt in enumerate(df['Player']):
    ax.annotate(txt,(df['Att 3rd'][i],df['90s'][i]))
```

```python
fig = plt.figure()
fig.set_size_inches(15,10)
ax = fig.add_subplot(1,1,1)
plt.style.use('seaborn-dark-palette')

#sns.stripplot(df['Mid 3rd'],df['90s'],jitter=0.25, size=8, ax=ax, linewidth=.5)
sns.scatterplot(x='%',y='Press',color='blue',data=df,marker='D',palette='deep',size='%',alpha=0.8)
plt.xlabel("Successful Pressure Percentage",size=16)
plt.ylabel("Number of Pressure Applied",size=16)
plt.suptitle("% Successfully Gained Possession",size=20,y=0.92)

for i,txt in enumerate(df['Player']):
    ax.annotate(txt,(df['%'][i],df['Press'][i]))
```
