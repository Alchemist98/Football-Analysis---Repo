```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("Table example - Premier league table.csv")
data.head()
```

```python
data = data.drop('Last 6',axis=1)
data.head()
```

```python
data
```

```python
# bar argument plots our data and its simplest it needs two arguments x,height.

plt.bar(x=np.arange(1,21),height=data['Pts'],width=0.8)
```

```python
plt.bar(x=np.arange(1,21),height=data['Pts'],width=0.8)
plt.title("Premier League 18/19")
plt.xticks(np.arange(1,21),data['Team'],rotation=90)
plt.xlabel("Teams")
plt.ylabel("Points")
plt.show()
```

```python
teamColours = ['#034694','#001C58','#5CBFEB','#D00027',
              '#EF0107','#DA020E','#274488','#ED1A3B',
               '#000000','#091453','#60223B','#0053A0',
               '#E03A3E','#1B458F','#000000','#53162f',
               '#FBEE23','#EF6610','#C92520','#BA1F1A']

plt.bar(x=np.arange(1,21),height=data['Pts'],width=0.8,color=teamColours)
plt.title("Premier League 18/19")
plt.xticks(np.arange(1,21),data['Team'],rotation=90)
plt.xlabel("Teams")
plt.ylabel("Points")
plt.show()
```
