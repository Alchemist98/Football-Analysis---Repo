```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("Table example - Premier league table.csv")
data = data.drop('Last 6',axis=1)
data
```

```python
data.head()
```

```python
plt.hlines(np.arange(1,21),xmin=0,xmax=data['Pts'],colors="skyblue")
plt.plot(data['Pts'], np.arange(1,21), "o")
plt.yticks(np.arange(1,21),data['Team'])
plt.title("18/19 PL Season ")
plt.xlabel("Pts")
plt.ylabel("Teams")
plt.show()
```

```python
teamColours = ['#034694','#001C58','#5CBFEB','#D00027',
              '#EF0107','#DA020E','#274488','#ED1A3B',
               '#000000','#091453','#60223B','#0053A0',
               '#E03A3E','#1B458F','#000000','#53162f',
               '#FBEE23','#EF6610','#C92520','#BA1F1A']

plt.hlines(np.arange(1,21),xmin=0,xmax=data['Pts'],colors=teamColours)
plt.plot(data['Pts'], np.arange(1,21), "o")
plt.yticks(np.arange(1,21),data['Team'])
plt.title("18/19 PL Season ")
plt.xlabel("Pts")
plt.ylabel("Teams")
plt.show()
```
