```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("championsleague.xlsx")
data.head()
```

```python
len(data)
```

```python
data.columns
```

```python
FW = data[data['Pos']=='FW']
FW.head()
```

```python
len(FW)
```

```python
df = FW[FW['90s']>=3]
```

```python
len(df)
```

```python
df.head(10)
```

```python
df.columns
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df = df.sort_values(by='xA',ascending=False)
```

```python
df
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df
```

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
plt.style.use("dark_background")

plt.scatter(df['xA'],df['Prog'],marker='D',color='blue')
plt.xlabel("Total Number of Assists per 90",size=16)
plt.ylabel("Number of progressive passes played per 90",size=16)

for i,txt in enumerate(df['Player']):
    ax.annotate(txt,(df['xA'][i],df['Prog'][i]))
```
