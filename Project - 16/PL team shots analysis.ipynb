```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("data.xlsx")
data
```

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
plt.style.use("dark_background")

plt.scatter(data['PenaltyArea'],data['Total'],marker='D',color='blue')
plt.xlabel("Shots taken in Penalty Area",size=16)
plt.ylabel("Total Shots taken",size=16)

for i,txt in enumerate(data['Team']):
    ax.annotate(txt,(data['PenaltyArea'][i],data['Total'][i]),size=15)
```

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
plt.style.use("dark_background")

plt.scatter(data['OutOfBox'],data['Total'],marker='D',color='blue')
plt.xlabel("Shots taken in Outside of Penalty Area",size=16)
plt.ylabel("Total Shots taken",size=16)

for i,txt in enumerate(data['Team']):
    ax.annotate(txt,(data['OutOfBox'][i],data['Total'][i]),size=15)
```

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
plt.style.use("dark_background")

plt.scatter(data['SixYardBox'],data['Total'],marker='D',color='blue')
plt.xlabel("Shots taken in Six Yard box Area",size=16)
plt.ylabel("Total Shots taken",size=16)

for i,txt in enumerate(data['Team']):
    ax.annotate(txt,(data['SixYardBox'][i],data['Total'][i]),size=15)
```

```python
df = data.sort_values(by='PenaltyArea',ascending=False).head(10)
df.reset_index(drop=True,inplace=True)
```

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)

sns.barplot(x='Team',y='PenaltyArea',data=df)
plt.title("Total numbers of shots taken in penalty area",size=18)
```
