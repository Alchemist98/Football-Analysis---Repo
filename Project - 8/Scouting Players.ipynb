```python
def player_replacement(player,skillReduction=0):
    
    replace_name = data[data['short_name'] == player][['short_name','wage_eur','age','overall','player_positions']]
    
    replace_age = replace_name['age'].item()
    replace_wages = replace_name['wage_eur'].item()
    replace_overall = replace_name['overall'].item()
    replace_pos = replace_name['player_positions'].item()
    
    
    longlist = data[data['player_positions'] == replace_pos][['short_name','wage_eur','age','overall','player_positions']]
    
    removals = longlist[longlist['age'] >= replace_age].index
    longlist.drop(removals,inplace=True)
    
    removals = longlist[longlist['wage_eur'] >= replace_wages].index
    longlist.drop(removals,inplace=True)
    
    removals = longlist[longlist['overall'] <= replace_overall].index
    longlist.drop(removals,inplace=True)
    
    return longlist.sort_values('wage_eur')
```

```python
player_replacement('M. Özil')
```
