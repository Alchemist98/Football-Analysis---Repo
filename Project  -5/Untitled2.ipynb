```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv("pokemon.csv")
data.head()
```

```python
data.columns
```

```python
import numpy as np
cols = np.array(['sp_attack','sp_defense','speed','attack','defense','capture_rate'])
stats = data.loc[11,cols].values
stats
```

```python
angles=np.linspace(0, 2*np.pi, len(cols), endpoint=False) # Set the angle
# close the plot
stats=np.concatenate((stats,[stats[0]]))  # Closed
angles=np.concatenate((angles,[angles[0]]))  
```

```python
fig=plt.figure()
ax = fig.add_subplot(111, polar=True)   # Set polar axis
ax.plot(angles, stats, 'o-', linewidth=2)  # Draw the plot (or the frame on the radar chart)
ax.fill(angles, stats, alpha=0.25)  #Fulfill the area
ax.set_thetagrids(angles * 180/np.pi, cols)  # Set the label for each axis
#ax.set_title([data.loc[386,"Name"]])  # Set the pokemon's name as the title
ax.set_rlim(0,100)
ax.grid(True)
```

```python
def plot_radars(x1,x2):
    labels = np.array(['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed'])
    
    stats1 = data.loc[x1,labels].values
    stats2 = data.loc[x2,labels].values
    
    angles = np.linspace(0,2*np.pi,len(labels),endpoint=False)
    
    stats1 = np.concatenate((stats1,[stats1[0]]))
    stats2 = np.concatenate((stats2,[stats2[0]]))
    angles = np.concatenate((angles,[angles[0]]))
    
    
    fig = plt.figure()
    ax = fig.add_subplot(111,polar=True)
    
    ax.plot(angles,stats1,stats2,'o-',linewidth=2)
    ax.fill(angles,stats1,stats2,alpha=0.25)
    ax.set_thetagrids(angles * 180/np.pi, cols)
    ax.set_rlim(0,250)
    ax.grid(True)
```

```python
plot_radars(111,100)
```
