```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_excel("playmakers.xlsx",index_col=0)
data
```

```python
def offensive_radars(x1,x2):
    labels = np.array(['Shots per game','Key Passes PG','Dribbles per Game','Crosses','Long Balls','Throw Balls'])
    
    stats1 = data.loc[x1,labels].values
    stats2 = data.loc[x2,labels].values
    
    angles = np.linspace(0,2*np.pi,len(labels),endpoint=False)
    angles2 = np.linspace(0,2*np.pi,len(labels),endpoint=False)
    
    stats1 = np.concatenate((stats1,[stats1[0]]))
    stats2 = np.concatenate((stats2,[stats2[0]]))
    angles = np.concatenate((angles,[angles[0]]))
    angles2 = np.concatenate((angles2,[angles2[0]]))
    
    fig = plt.figure()
    ax = fig.add_subplot(111,polar=True)
    
    plt.xticks(angles[:-1],labels)

    ax.plot(angles,stats1)
    ax.fill(angles, stats1, 'teal', alpha=0.1)

    ax.plot(angles2,stats2)
    ax.fill(angles2, stats2, 'red', alpha=0.1)

    #Rather than use a title, individual text points are added
    plt.figtext(0.2,0.9,"Kevin de bruyne",color="teal")
    plt.figtext(0.2,0.85,"v")
    plt.figtext(0.2,0.8,"Jack Grealish",color="red")
    plt.show()
```

```python
offensive_radars(1,6)
```
