```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.patches import Arc
import seaborn as sns
```

```python
data = pd.read_excel("Passing.xlsx")
data.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
data.head()
```

```python
fig = plt.figure()
ax = fig.set_size_inches(14,4)

plt.subplot(121)
sns.kdeplot(data["XSTART"],data["YSTART"],shade=True,kernel='tri')

plt.subplot(122)
sns.kdeplot(data["XSTART"],data["YSTART"])
plt.show()
```

```python
fig = plt.figure()
ax = fig.set_size_inches(14,4)

plt.subplot(121)
sns.kdeplot(data["XSTART"],data["YSTART"],shade=True,n_levels=5)

plt.subplot(122)
sns.kdeplot(data["XSTART"],data["YSTART"],shade=True,n_levels=40)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()

#Pitch
plt.plot([0,0],[0,120],color='black')
plt.plot([0,150],[120,120],color='black')
plt.plot([150,150],[120,0],color='black')
plt.plot([0,150],[0,0],color='black')
plt.plot([75,75],[0,120],color='black')

#Left Penalty Box
plt.plot([0,25],[30,30],color='black')
plt.plot([0,25],[90,90],color='black')
plt.plot([25,25],[30,90],color='black')

#Six Yard box
plt.plot([0,10],[45,45],color='black')
plt.plot([0,10],[75,75],color='black')
plt.plot([10,10],[45,75],color='black')

#Right Penalty Box
plt.plot([150,125],[30,30],color='black')
plt.plot([150,125],[90,90],color='black')
plt.plot([125,125],[30,90],color='black')

#Six Yard box
plt.plot([150,140],[45,45],color='black')
plt.plot([150,140],[75,75],color='black')
plt.plot([140,140],[45,75],color='black')

#Circles
Centrecircle = plt.Circle((75,60),radius=15.5,color='black',fill=False)
CentreSpot = plt.Circle((75,60),radius=1,color='black',fill=True)
LeftPenSpot = plt.Circle((18,60),radius=1,color='black',fill=True)
RightPenSpot = plt.Circle((132,60),radius=1,color='black',fill=True)

#Arcs
leftArc = Arc((13,60),height=38.3,width=38.3,angle=0,theta1=310,theta2=50,color="black")
rightArc = Arc((137,60),height=38.3,width=38.3,angle=0,theta1=130,theta2=230,color="black")

ax.add_patch(Centrecircle)
ax.add_patch(CentreSpot)
ax.add_patch(LeftPenSpot)
ax.add_patch(RightPenSpot)
ax.add_patch(leftArc)
ax.add_patch(rightArc)


#Tidy Axes
plt.axis('off')

sns.kdeplot(data["XSTART"],data["YSTART"], shade=True,n_levels=50)
plt.ylim(0,150)
plt.xlim(0,180)

plt.show()
```
