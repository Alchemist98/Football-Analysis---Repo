```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.patches import Arc 
```

```python
data = pd.read_csv("Passes.csv")
data.head()
```

```python
data = data.drop('Unnamed: 0',axis=1)
data
```

```python
#We have time given in addition to starting and ending of x and y.
#We can .plot() func to draw line if we give it two lists
#List one must contain the start and end X locations
#List two gives the start and end Y locations

#For example, plt.plot([0,1],[2,3] will plot a line from location (0,2) to (1,3)

fig, ax = plt.subplots()

fig.set_size_inches(7,5)


for i in range(len(data)):
    plt.plot([int(data['XSTART'][i]),int(data['XEND'][i])],
            [int(data['YSTART'][i]),int(data['YEND'][i])],
            color="green")
    
plt.show()
```

```python
fig, ax = plt.subplots()

fig.set_size_inches(7,5)


for i in range(len(data)):
    plt.plot([int(data['XSTART'][i]),int(data['XEND'][i])],
            [int(data['YSTART'][i]),int(data['YEND'][i])],
            color="green")
    
    plt.plot([int(data['XSTART'][i])],[int(data['YSTART'][i])],"o",color="blue")
    plt.plot([int(data['XEND'][i])],[int(data['YEND'][i])],"o",color="blue")
    
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(7,5)

ax = fig.add_subplot(1,1,1)

#Add outer lines and centre lines
plt.plot([0,0],[0,90],color="black")
plt.plot([0,130],[90,90],color="black")
plt.plot([0,130],[0,0],color="black")
plt.plot([130,130],[0,90],color="black")
plt.plot([65,65],[0,90],color="black")

#Left Pen Area
plt.plot([0,15],[30,30],color="black")
plt.plot([0,15],[60,60],color="black")
plt.plot([15,15],[30,60],color="black")

#Left Six-Yard box
plt.plot([0,5],[40,40],color="black")
plt.plot([0,5],[50,50],color="black")
plt.plot([5,5],[40,50],color="black")

#Left Pen Spot
LeftPenSpot = plt.Circle((10,45),radius=1.0,color="grey")

#Right Pen Area
plt.plot([130,115],[30,30],color="black")
plt.plot([130,115],[60,60],color="black")
plt.plot([115,115],[30,60],color="black")

#Right Six-Yard box
plt.plot([130,125],[40,40],color="black")
plt.plot([130,125],[50,50],color="black")
plt.plot([125,125],[40,50],color="black")

#Right Pen Spot
RightPenSpot = plt.Circle((120,45),radius=1.0,color="grey")

#Prepare Circles
CentreCircle = plt.Circle((65,45),radius=9.5,color="red",fill=False)
CentreSpot = plt.Circle((65,45),radius=1.0,color="red")

#Create Arcs
LeftArc = Arc((9,45),width=18.3,height=18.3,angle=0,theta1=310,theta2=50,color="black")
RightArc = Arc((121,45),width=18.3,height=18.3,angle=0,theta1=130,theta2=230,color="black")

#Add circles
ax.add_patch(CentreCircle)
ax.add_patch(CentreSpot)
ax.add_patch(LeftPenSpot)
ax.add_patch(RightPenSpot)
ax.add_patch(LeftArc)
ax.add_patch(RightArc)

for i in range(len(data)):
    plt.plot([int(data['XSTART'][i]),int(data['XEND'][i])],
            [int(data['YSTART'][i]),int(data['YEND'][i])],
            color="green")
    
    plt.plot([int(data['XSTART'][i])],[int(data['YSTART'][i])],"o",color="blue")
    plt.plot([int(data['XEND'][i])],[int(data['YEND'][i])],"o",color="blue")


plt.title("Mesut Ozil Passes in 1st Half")
plt.show()
```
