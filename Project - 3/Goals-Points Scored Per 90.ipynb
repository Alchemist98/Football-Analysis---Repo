```python
def p90_calculator(var_value,min_played):
    
    ninety_min_rounds = min_played/90
    
    goals_scored_p90 = var_value/ninety_min_rounds
    
    return goals_scored_p90

p90_calculator(25,2111)
```

```python
def p90_Calculator(value, minutes):
    return value/(minutes/90)

p90_Calculator(20,1835)
```

```python
data['total_pts_p90'] = p90_calculator(data.Points,data.Minutes)
data['total_pts_p90'].fillna(0,inplace=True)
data
```
